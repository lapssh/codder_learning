### Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
full_data = ratings.merge(movies, on = 'movieId', how='left')
movies_by_rating_5 = full_data.query('rating == 5.0')
best_film_index = movies_by_rating_5['movieId'].value_counts().head(1).index[0]
print('Самое большое число оценок 5.0 выставлено фильму: ')
movies[movies['movieId'] == best_film_index]
```

### Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power = pd.read_csv('power.csv')
power.head(3)
```

```python
# отсекаем только страны Прибалтики
baltic = power[(power['country'] == 'Latvia') | (power['country'] == 'Lithuania') | (power['country'] == 'Estonia')]
```

```python
# отскекаем с 2005 по 2010 гг.
baltic2 = baltic[(baltic['year'] >= 2005) & (baltic['year'] <= 2010)]
baltic2.head()
```

```python
# отсекаем категории 4, 12 и 21
baltic3 = baltic2[(baltic2['category'] == 4)|(baltic2['category'] == 12)|(baltic2['category'] == 21)]
baltic3.head()
```

```python
# оставляем значения с неотрицательным потреблением
baltic4 = baltic3[baltic3['quantity'] >= 0]
baltic4.head()
```

```python
# Наконец считаем суммарное потребление
baltic_quantity = sum(baltic4['quantity'])
print('Суммарное потребление Прибалтики за период 2005-2010 гг: ',baltic_quantity)
```
